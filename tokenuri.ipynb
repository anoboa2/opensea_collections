```python
for index, collection in collections.iterrows():
    if 'tokenURI' in set(checkTokenURI(collection['contract_address'])['name']):
        print(f'SUCCESS:\t{collection.slug} is using the EIP721 metadata extension standard')
        data, headers = getTokens(collection['contract_address'], collection['slug'], collection['total_supply'])
        print(f'Finished checking all tokens for {collection.slug}')
        output = pd.DataFrame(data=data, columns=headers)
        output.to_csv(f'tokens/{collection.slug}_tokens.csv')
        print(f'Successfully wrote {collection.slug} results to csv file')
    else:
        print(f'ERROR:\t{collection.slug} is NOT using the EIP721 metadata extension standard')
```
